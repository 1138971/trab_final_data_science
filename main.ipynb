```python
dados.rename(columns={"Por quais motivos você escolheu cursar CC?": "principal_motivo_escolha"}, inplace=True)
dados.columns
```

```python
dados["principal_motivo_escolha"] = dados["principal_motivo_escolha"].str.lower().str.strip()
dados.head()
```

```python
substituicoes = {
    "interesse por tecnologia": "Interesse pessoal",
    "alta empregabilidade": "Mercado de trabalho / salário",
    "possibilidade de trabalho remoto": "Mercado de trabalho / salário",
    "prestígio da área no mercado": "Mercado de trabalho / salário",
    "promessa de bons salários": "Mercado de trabalho / salário",
    "alta empregabilidade": "Mercado de trabalho / salário",
    "influência de familiares ou amigos": "Influência de terceiros",
    "influencia de familiares ou amigos": "Influência de terceiros"
}

for antigo, novo in substituicoes.items():
    dados["principal_motivo_escolha"] = dados["principal_motivo_escolha"].str.replace(antigo, novo, regex=False)

```
